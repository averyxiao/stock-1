```python
import sys
sys.path.append('..')
from configure.settings import DBSelector
```

```python
db=DBSelector().mongo('qq')
```

```python
doc=db['db_stock']['danjuan_fund']
```

```python
plan_list =[]
for item in doc.find():
    plan_name=item['plan_name']
    holding=item['holding']
    d={}
    d[plan_name]=holding
    plan_list.append(d)
```

```python
len(plan_list)
```

```python
plan_list[0]
```

```python
fund_dict={}
for item in plan_list:
    name=list(item.keys())[0]
    funds=item[name]
    for fund in funds:
        # if fund['percent']!=0:
        fund_dict.setdefault(fund['fd_name'],set())
        fund_dict[fund['fd_name']].add(name)
```

```python
result={}
for item in fund_dict:
    print(item,len(fund_dict[item]))
    result[item]=len(fund_dict[item])
```

```python
list(sorted(result.items(),key=lambda x:x[1],reverse=True))
```

```python
help(sorted)
```

```python
result_=list(sorted(result.items(),key=lambda x:x[1],reverse=True))
```

```python
result_dict_=[]
for name,count in result_:
    d={}
    d['name']=name
    d['count']=count
    result_dict_.append(d)
```

```python
import pandas as pd
```

```python
result_dict_
```

```python
df = pd.DataFrame(result_dict_)
```

```python
df.head()
```

```python
df.to_excel('fund_count_.xls',encoding='utf8')
```
