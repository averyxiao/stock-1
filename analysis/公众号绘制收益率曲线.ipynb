```python
with open(path,'r',encoding='utf8') as fp:
    content = fp.readlines()
```

```python
len(content)
```

```python
content
```

```python
import re

result=[]
for line in content:
    line=line.strip()
    if line:
#         print(line)
        m=re.search('(.*?) 00:00:00 .*?收益率 ： (.*?)% ',line)
        if m:
            date=m.group(1)
            profit=m.group(2)
            result.append({'date':date,'profit':profit})
            
```
