```python
engine = DBSelector().get_engine('db_closed_end_daily')
```

```python
excel_file = '/home/xda/hub/stock/data/closed_end_data.xlsx'
```

```python
import pandas as pd
```

```python
df = pd.read_excel(excel_file)
```

```python
df.head()
```

```python
df['code']=df['code'].map(lambda x:str(x))
```

```python
codes = df['code'].values
```

```python
for code in codes:
    for _ in range(5):
        if code.startswith('1'):
            code=code+'.SZ'
        else:
            code=code+'.SH'
        try:
            fund_netvalue  = pro.fund_nav(ts_code=code)
            fund_netvalue.to_sql(code,con=engine,if_exists='replace')
        except Exception as e:
            print(e)
        else:
            break
```
